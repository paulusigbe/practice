```python
#USING DICTIONARY OF LIST FOR DATA FRAME
#TAKE HOME PRACTICE
import pandas as pd

sales = { 
'name':["Paul", "Peter", "James", "John", "Mark"],
'age':[25, 30, 40, 20, 35],
'city':["Lagos", "Abuja", "Kano", "Jos", "Kanduna"]
}

df = pd.DataFrame(sales)
#PRINTING ONLY THE NAME COLUMN

print(df[['name']])
print()
print()

#SORT DATA FRAME FROM THE HIGHEST AGE

sortdf = df.sort_values(by = 'age', ascending = False)

print(sortdf)
print()
print()

#SORT DATA FRAME FOR THE HIGHEST

maxAge = df['age'].max()

print(maxAge)
print()

#READ AS CSV
#FIRST SAVE TO A DIRECTORY
df.to_csv(r"/Users/user/Desktop/dataframe.csv", index=False) 

print("Successfully saved on the Desktop")
print()
print()

#TO READ AS CSV, 
#THE read_csv() FUNCTION IS TO READ DATA FROM CSV FILES IN PANDA
dfs = pd.read_csv("/Users/user/Desktop/dataframe.csv")

print(dfs)
print()

df.head()
```

```python
#CREATE A DATAFRAME TABLE WITH THE FOLLOWING COLUMNS, ORDER_ID, PRODUCT, PRICE, QUANTITY, REGION.
#USE THE MOST COMMON DATAFRAME. DICTIONARY OF LIST
import pandas as pd

sale_Shop = {
'Order_Id':[10001, 10002, 10003, 10004, 100005], 
'Product':['Shoe', 'Belt', 'Shirt', 'Hat', 'Sandals'], 
'Price':[50000, 20000, 25000, 15000, 30000], 
'Quantity':[6, 3, 10, 2, 4], 
'Region':['Kaduna', 'Lagos', 'Abuja', 'Kano', 'Rivers']

            }

df = pd.DataFrame(sale_Shop)

print(df)
print()
print()

df.head()
df.info()
df.describe()
```
